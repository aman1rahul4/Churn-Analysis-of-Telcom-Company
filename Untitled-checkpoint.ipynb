```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0")# to replace blank space to 0
df["TotalCharges"]= df["TotalCharges"].astype("float") # to convert date type from object to float
```

#to convert 0 and 1 values of SeniorCitizen columns into yes and no

```python
def conv(value):
    if value == 1:
        return"yes"
    else :
        return"no"
df['SeniorCitizen']= df["SeniorCitizen"].apply(conv)
```

```python
df.head(70)
```

```python
ax = sns.countplot(x= 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customer  by Churn")
plt.show()
```

#Above bar garph explains the number of churn  

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index,autopct = '%1.2f%%')
plt.title("Churn Percent")
plt.show()
```

#Above pie chart concludes that 26.54% customers has left services and 73.46% customers are still using services.

```python
plt.figure(figsize=(3,3))
sns.countplot( x ='gender', data = df, hue ="Churn",)
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot( x ='SeniorCitizen', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by SeniorCitizen")
plt.show()
```

```python
# Calculate percentage
df_grouped = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
df_total = df_grouped.groupby('SeniorCitizen')['Count'].transform('sum')
df_grouped['Percentage'] = df_grouped['Count'] / df_total *100
# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(4,4))
df_pivot = df_grouped.pivot(index='SeniorCitizen', columns='Churn', values='Percentage')
df_pivot.plot(kind='bar', stacked=True, ax=ax)
# Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', label_type='center')
plt.title("Churn by SeniorCitizen (Percentage)")
plt.ylabel("Percentage")
plt.show()
```

```python

# List of columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create a figure with subplots
num_cols = len(columns)
num_rows = (num_cols // 3) + (num_cols % 3 > 0)  # Adjust rows depending on number of columns

# Set up the figure size
plt.figure(figsize=(15, 5 * num_rows))

# Loop through each column and create a count plot
for i, col in enumerate(columns):
    plt.subplot(num_rows, 3, i + 1)  # Create subplots (3 per row)
    sns.countplot(x=col, data=df, hue= df["Churn"])
    plt.title(f'Count Plot for {col}')
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

#The subplot grid created using Seaborn visualizes the count distribution of various services (such as PhoneService, MultipleLines, InternetService, and others) in your dataset. 
Each subplot shows the number of instances for each category within a service, giving you a clear overview of your data's composition. 
This layout helps in quickly comparing different features side-by-side.
